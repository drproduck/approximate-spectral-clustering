```matlab
clear

datasets = {'pend', 'usps', 'mnist'};
id = fopen('/srv/home/009984221/dump','w');
rng(1234);

alpha = 0.01;

n_iteration = 20;
accuracies = zeros([3 4 n_iteration]);
time = zeros([3 4 n_iteration]);

%% experiments

fprintf('\n***NJW spectral clustering on digits data (Table II)***\n')

for i = 1:length(datasets)
    fprintf('\nloading %s...\n', datasets{i})
    load(datasets{i})
    A = fea;
    y = gnd;
    y(y==0)=max(y)+1;

    n = numel(y);
    k = max(y);
    
    for j = 1:n_iteration
        fprintf('\niteration %d...\n', j)

        % use same sigma for all algorithms
        sigma = estimate_sigma_knn(fea, 7, 100);

        fprintf('running scalable NJW...')
        t0 = cputime;
        inds_sc = ssc_cosine(A, k, struct('t', -1, 'alpha', alpha));
        runtime = cputime - t0;
        inds_sc = bestMap(y,inds_sc);
        accuracy = sum(y == inds_sc)/length(y);
        fprintf('accuracy=%0.4f, cputime=%2.1f\n', accuracy, runtime)
        accuracies(i, 1, j) = accuracy;
        time(i, 1, j) = runtime;

        fprintf('running scalable NJW with gaussian embedding...')
        t0 = cputime;
        inds_sc = ssc_gaussian(A, k, 500, 5, struct('t', -1, 'alpha', alpha, 'sigma', sigma));
        runtime = cputime - t0;
        inds_sc = bestMap(y,inds_sc);
        accuracy = sum(y == inds_sc)/length(y);
        fprintf('accuracy=%0.4f, cputime=%2.1f\n', accuracy, runtime)
        accuracies(i, 2, j) = accuracy;
        time(i, 2, j) = runtime;

        fprintf('running LBDM...')
        t0 = cputime;
        inds_sc = LBDM(A, k, 500, 5, 2, 'gaussian','fileid',id, 'sigma', sigma);
        runtime = cputime - t0;
        inds_sc = bestMap(y,inds_sc);
        accuracy = sum(y == inds_sc)/length(y);
        fprintf('accuracy=%0.4f, cputime=%2.1f\n', accuracy, runtime)
        accuracies(i, 3, j) = accuracy;
        time(i, 3, j) = runtime;

        fprintf('running LSC ...')
        t0 = cputime;
        inds_sc = LSC(A, k, struct('p', 500, 'r', 5, 'mode', 'kmeans', 'sigma', sigma));
        runtime = cputime - t0;
        inds_sc = bestMap(y,inds_sc);
        accuracy = sum(y == inds_sc)/length(y);
        fprintf('accuracy=%0.4f, cputime=%2.1f\n', accuracy, runtime)
        accuracies(i, 4, j) = accuracy;
        time(i, 4, j) = runtime;
    
    
    %     fprintf('running plain NJW...')
    %     t0 = cputime;
    %     inds_sc = specluster(A, k, struct('t', -1, 'affinity', 'cosine', 'alpha', alpha));
    %     runtime = cputime - t0;
    %     inds_sc = bestMap(y,inds_sc);
    %     accuracy = sum(y == inds_sc)/length(y);
    %     fprintf('accuracy=%2.2f, cputime=%2.1f\n', accuracy, runtime)

    end
    
end
```
